    - Usa un `if` para verificar si la `enfermedad` ya está en `lista_enfermedades`.
    - Si **no** está en `lista_enfermedades`, agrega la enfermedad a `lista_enfermedades` y añade `1` a `lista_cantidades` para contarla por primera vez.
    - Si **sí** está en `lista_enfermedades`, encuentra la posición (`indice`) de la enfermedad en `lista_enfermedades` y aumenta en `1` el valor en `lista_cantidades` en la misma posición.

````python
# Pista: Usa `in` para verificar si la enfermedad ya está en lista_enfermedades.
# Pista: Si la enfermedad no está, agrégala a lista_enfermedades y empieza el conteo en lista_cantidades.
# Pista: Si la enfermedad ya está, usa `index` para encontrar su posición en lista_enfermedades y suma 1 en lista_cantidades en esa posición.
lista_enfermedades = []
lista_cantidades = []           
datos_pacientes = clinica_intermedio + clinica

for paciente in datos_pacientes:
    if isinstance(paciente, list) and len(paciente) > 4: 
        enfermedad = paciente[4]
        if enfermedad not in lista_enfermedades:
            lista_enfermedades.append(enfermedad)
            lista_cantidades.append(1)
        else:
            indice = lista_enfermedades.index(enfermedad)
            lista_cantidades[indice] += 1

            print("\nLista de Enfermedades")
            print(lista_enfermedades)
            print("\nLista de Cantidades")
            print([f"{enfermedad}: {cantidad}" for enfermedad, cantidad in zip(lista_enfermedades, lista_cantidades)])     
````

#### **Paso 4: Imprimir el Reporte Final**

1. **Objetivo:** Mostrar el número de pacientes con cada enfermedad.
    
2. **Instrucción:**
    
    - Usa un bucle `for` para recorrer la lista `lista_enfermedades`.
    - En cada iteración, imprime la enfermedad y el número de pacientes (que está en `lista_cantidades` en la misma posición).

````python
lista_enfermedades = []
lista_cantidades = []

datos_pacientes = clinica_intermedio + clinica

contador_enfermedades = {}


for paciente in datos_pacientes:
    if isinstance(paciente, list) and len(paciente) > 4:  
        enfermedad = paciente[4]  
        if enfermedad not in lista_enfermedades:
            lista_enfermedades.append(enfermedad) 
            lista_cantidades.append(1)  
        else:
            indice = lista_enfermedades.index(enfermedad)
            lista_cantidades[indice] += 1

print("\nReporte de Enfermedades Tratadas:")
for enfermedad, cantidad in zip(lista_enfermedades, lista_cantidades):
    print(f"Enfermedad: {enfermedad} - Pacientes: {cantidad}")
````

````python
import os
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ruta_clinica_intermedio = r"C:\Users\Usuario\Documents\Curso de Python\clinica_intermedio.json"
ruta_clinica = r"C:\Users\Usuario\Documents\Curso de Python\clinica.json"

ruta_clinica_intermedio = r"C:\Users\Usuario\Documents\Curso de Python\clinica_intermedio.json"
ruta_clinica = r"C:\Users\Usuario\Documents\Curso de Python\clinica.json"

# Cargar los archivos
if os.path.exists(ruta_clinica_intermedio) and os.path.exists(ruta_clinica):
    print("Los archivos existen. Intentando abrirlos...")
    try:
        with open(ruta_clinica_intermedio, encoding='utf-8') as archivo1:
            clinica_intermedio = json.load(archivo1)
        with open(ruta_clinica, encoding='utf-8') as archivo2:
            clinica = json.load(archivo2)      
        print(f"Se han cargado {len(clinica_intermedio)} registros de 'clinica_intermedio'.")
        print(f"Se han cargado {len(clinica)} registros de 'clinica'.")
    except json.JSONDecodeError:
        print("Error: El archivo JSON tiene un formato inválido. Revisa su contenido.")
else:
    print("Error: Uno o ambos archivos no existen en la ubicación esperada.")

# Unir los datos de los pacientes
datos_pacientes = clinica_intermedio + clinica

# Función para obtener enfermedades y medicamentos
def obtener_enfermedades_y_medicamentos(paciente):
    enfermedad = paciente[4] if len(paciente) > 4 else None
    medicamento = paciente[5] if len(paciente) > 5 else None
    return enfermedad, medicamento

# Inicializar listas para almacenar enfermedades y medicamentos comunes
enfermedades_comunes = []
medicamentos_comunes = []

# Comparar todos los pacientes
for i, paciente_1 in enumerate(datos_pacientes):
    for j, paciente_2 in enumerate(datos_pacientes):
        if i >= j:
            continue  # Evitar comparaciones repetidas o consigo mismo
        
        enfermedad_1, medicamento_1 = obtener_enfermedades_y_medicamentos(paciente_1)
        enfermedad_2, medicamento_2 = obtener_enfermedades_y_medicamentos(paciente_2)

        # Listas para almacenar las enfermedades y medicamentos comunes
        if enfermedad_1 and enfermedad_2 and enfermedad_1 == enfermedad_2:
            enfermedades_comunes.append(enfermedad_1)
        
        if medicamento_1 and medicamento_2 and medicamento_1 == medicamento_2:
            medicamentos_comunes.append(medicamento_1)

# Contar las ocurrencias de enfermedades y medicamentos comunes
enfermedades_contadas = pd.Series(enfermedades_comunes).value_counts()
medicamentos_contados = pd.Series(medicamentos_comunes).value_counts()

# Crear DataFrames para las visualizaciones
df_enfermedades = pd.DataFrame({'Enfermedad': enfermedades_contadas.index, 'Pacientes': enfermedades_contadas.values})
df_medicamentos = pd.DataFrame({'Medicamento': medicamentos_contados.index, 'Pacientes': medicamentos_contados.values})

# Crear gráficos

# Gráfico de enfermedades comunes
plt.figure(figsize=(10, 6))
sns.barplot(x='Enfermedad', y='Pacientes', data=df_enfermedades, palette='Blues_d')
plt.title('Enfermedades más comunes entre pacientes', fontsize=16)
plt.xlabel('Enfermedad', fontsize=12)
plt.ylabel('Número de Pacientes', fontsize=12)
plt.xticks(rotation=45)  # Rotar las etiquetas para que no se superpongan
plt.tight_layout()  # Asegurarse de que todo esté bien acomodado
plt.show()

# Gráfico de medicamentos comunes
plt.figure(figsize=(10, 6))
sns.barplot(x='Medicamento', y='Pacientes', data=df_medicamentos, palette='Greens_d')
plt.title('Medicamentos más comunes entre pacientes', fontsize=16)
plt.xlabel('Medicamento', fontsize=12)
plt.ylabel('Número de Pacientes', fontsize=12)
plt.xticks(rotation=45)  # Rotar las etiquetas para que no se superpongan
plt.tight_layout()  # Asegurarse de que todo esté bien acomodado
plt.show()
````
